## Importing Libraries

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

## Loading dataset 

```python
df = load_iris()
```

## Summarize the dataset 

```python
df
```

## Segregate the dataset into X & Y 

```python
X = pd.DataFrame(df.data,columns=df.feature_names)
X
```

```python
Y = df.target
Y
```

So let clear the this 0 , 1 ,2 what does the represent which class

### 0 = setosa , 1 = versicolor , 2 = virginica

## Training with train_test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state = 0)
print(X_train.shape)
print(X_test.shape)
```

## Finding the Best max_depth Value

```python
accuracy = []
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

for i in range(1,10):
    model = DecisionTreeClassifier (max_depth = i , random_state = 0)
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    score = accuracy_score(y_test,pred)
    accuracy.append(score)

plt.figure(figsize=(12,6))
plt.plot(range(1,10),accuracy,color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.title('Finding the Best Max Depth')
plt.xlabel('pred')
plt.ylabel('score')
```

## Training the dataset

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'entropy',max_depth = 3 , random_state = 0)
model.fit(X_train,y_train)
```

## Prediction

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

## Accuracy Score

```python
from sklearn.metrics import accuracy_score
print("the Accuracy Score this model is :{0}%".format(accuracy_score(y_test,y_pred)*100))
```
